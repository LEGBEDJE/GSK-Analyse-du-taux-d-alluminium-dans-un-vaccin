```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
# charger les données
from google.colab import drive

```

```python
# Acceder au différentes feuilles
ex1_data = data['Ex1']
ex2_data = data['Ex2']
ex1_data
ex2_data
```

```python
from matplotlib import pyplot as plt
ex1_data['Aluminum Content (mg/mL)'].plot(kind='hist', bins=20, title='Aluminum Content (mg/mL)')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# Statistiques descriptives
print(ex1_data['Aluminum Content (mg/mL)'].describe())
```

```python
# Visualisation de la tendance temporelle
plt.figure(figsize=(10,6))
sns.lineplot(data=ex1_data, x='Batch number (Chronological Order)', y='Aluminum Content (mg/mL)')

plt.axhline(1, color='red', linestyle='--')  # Valeur cible
plt.axhline(0.8, color='green', linestyle='--')  # Limite inférieure
plt.axhline(1.2, color='green', linestyle='--')  # Limite supérieure
plt.title('Tendance du Contenu en Aluminium (2018-2022)')
plt.show()
```

```python
# Diagramme de contrôle
plt.figure(figsize=(10,6))
sns.lineplot(data=ex1_data, x='Batch number (Chronological Order)', y='Aluminum Content (mg/mL)', marker='o')
plt.axhline(ex1_data['Aluminum Content (mg/mL)'].mean(), color='blue', linestyle='--')
plt.axhline(ex1_data['Aluminum Content (mg/mL)'].mean() + 3*ex1_data['Aluminum Content (mg/mL)'].std(), color='red', linestyle='--')
plt.axhline(ex1_data['Aluminum Content (mg/mL)'].mean() - 3*ex1_data['Aluminum Content (mg/mL)'].std(), color='red', linestyle='--')
plt.title('Diagramme de Contrôle du Contenu en Aluminium')
plt.show()

```

```python
# recherche approfondie sur les donées
import scipy.stats as stats

# Histogramme et courbe de densité
plt.figure(figsize=(10,6))
sns.histplot(ex1_data['Aluminum Content (mg/mL)'], kde=True, bins=30)
plt.axvline(1, color='red', linestyle='--', label='Valeur cible')
plt.title('Distribution du Contenu en Aluminium')
plt.xlabel('Contenu en Aluminium (mg/mL)')
plt.ylabel('Fréquence')
plt.legend()
plt.show()

# Test de normalité (Shapiro-Wilk)
stat, p = stats.shapiro(ex1_data['Aluminum Content (mg/mL)'])
print(f'Test de Shapiro-Wilk: Statistique={stat}, p-valeur={p}')

```

```python
# Boxplot pour détecter les outliers
plt.figure(figsize=(10,6))
sns.boxplot(x=ex1_data['Aluminum Content (mg/mL)'])
plt.axvline(1, color='red', linestyle='--', label='Valeur cible')
plt.title('Boxplot du Contenu en Aluminium')
plt.xlabel('Contenu en Aluminium (mg/mL)')
plt.legend()
plt.show()
# Identification des outliers
Q1 = ex1_data['Aluminum Content (mg/mL)'].quantile(0.25)
Q3 = ex1_data['Aluminum Content (mg/mL)'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = ex1_data[(ex1_data['Aluminum Content (mg/mL)'] < lower_bound) | (ex1_data['Aluminum Content (mg/mL)'] > upper_bound)]
print('Outliers détectés :')
print(outliers)
```

```python
# Calcul du taux de non-conformité
total_batches = len(ex1_data)
non_conform_batches = len(ex1_data[(ex1_data['Aluminum Content (mg/mL)'] < 0.8) | (ex1_data['Aluminum Content (mg/mL)'] > 1.2)])
non_conform_rate = (non_conform_batches / total_batches) * 100
print(f'Taux de non-conformité: {non_conform_rate:.2f}%')



```

```python
# Calcul de Cp et Cpk
spec_lower = 0.8
spec_upper = 1.2
process_mean = ex1_data['Aluminum Content (mg/mL)'].mean()
process_std = ex1_data['Aluminum Content (mg/mL)'].std()

Cp = (spec_upper - spec_lower) / (6 * process_std)
Cpk = min((spec_upper - process_mean) / (3 * process_std), (process_mean - spec_lower) / (3 * process_std))

print(f'Cp: {Cp:.2f}, Cpk: {Cpk:.2f}')

```

```python
# Analyse de l'Exercise 2

# Calculer la différence des paires
ex2_data['Difference'] = ex2_data['siteA'] - ex2_data['siteB']

# Statistiques descriptives
print(ex2_data['Difference'].describe())

# Test de t pour échantillons appariés
t_stat, p_val = stats.ttest_rel(ex2_data['siteA'], ex2_data['siteB'])
print(f"t-statistique: {t_stat}, p-valeur: {p_val}")

# Vérification de l'équivalence
mean_diff = ex2_data['Difference'].mean()
if -0.1 < mean_diff < 0.1:
    print("Les sites A et B sont équivalents.")
else:
    print("Les sites A et B ne sont pas équivalents.")
```

```python
# Distribution des différences
plt.figure(figsize=(10, 6))
sns.histplot(ex2_data['Difference'], kde=True, bins=30)
plt.axvline(ex2_data['Difference'].mean(), color='r', linestyle='--', label='Moyenne')
plt.title('Distribution des Différences de Contenu en Aluminium entre les Sites A et B')
plt.xlabel('Différence (mg/mL)')
plt.ylabel('Fréquence')
plt.legend()
plt.show()

# Visualisation des moyennes avec intervalles de confiance
mean_diff = ex2_data['Difference'].mean()
ci = stats.t.interval(0.95, len(ex2_data['Difference'])-1, loc=mean_diff, scale=stats.sem(ex2_data['Difference']))

plt.figure(figsize=(8, 6))
plt.errorbar(x=['Différence Moyenne'], y=[mean_diff], yerr=[[mean_diff - ci[0]], [ci[1] - mean_diff]], fmt='o', capsize=5, capthick=2)
plt.axhline(0.1, color='g', linestyle='--', label='Limite Supérieure d\'Équivalence')
plt.axhline(-0.1, color='g', linestyle='--', label='Limite Inférieure d\'Équivalence')
plt.axhline(0, color='r', linestyle='-', label='Équivalence Parfaite')
plt.title('Visualisation de la Différence Moyenne avec Intervalles de Confiance')
plt.ylabel('Différence Moyenne (mg/mL)')
plt.legend()
plt.show()
```

```python
# Analyse de la Variabilité
std_siteA = ex2_data['siteA'].std()
std_siteB = ex2_data['siteB'].std()
print(f"Écart-type Site A: {std_siteA}")
print(f"Écart-type Site B: {std_siteB}")

# Analyse de la Corrélation
correlation = ex2_data['siteA'].corr(ex2_data['siteB'])
print(f"Corrélation entre Site A et Site B: {correlation}")

# Tests de Normalité
from scipy.stats import shapiro

stat, p = shapiro(ex2_data['Difference'])
print(f"Shapiro-Wilk test statistique: {stat}, p-valeur: {p}")
```
